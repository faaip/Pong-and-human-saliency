```python
import torch, glob, re, os, pickle
import numpy as np

import torch.nn as nn
from torch.utils.data.dataset import Dataset
from torch.utils.data.sampler import SubsetRandomSampler
from torch.utils.data import DataLoader

import torch.optim as optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
batch_size = 100
learn_rate = 0.00000001
train_test_split = .01
random_seed = 42
```

```python
class DriveData(Dataset):
  __xs = []
  __ys = []

  def __init__(self, folder_dataset):
    # Open and load text file including the whole training data
    data_parts = glob.glob(os.path.join(folder_dataset,'*'))
    for file in data_parts:
      with open(file,'rb') as f:
        data, label = pickle.load(f)
        for i in range(len(data)):
          self.__xs.append(data[i])
          self.__ys.append(label[i][np.newaxis])
                               
  # Override to give PyTorch access to any image on the dataset
  def __getitem__(self, index):
    img = torch.FloatTensor(self.__xs[index])
    label = torch.FloatTensor(self.__ys[index])
    return img, label

  # Override to give PyTorch size of dataset
  def __len__(self):
    return len(self.__xs)
  
```

```python
data_directory = '/Users/maximoldenbeek/Dropbox/Saliency_dataset/'
dataset = DriveData(data_directory)

# Creating data indices for training and validation splits:
dataset_size = len(dataset)
indices = list(range(dataset_size))
split = int(np.floor(train_test_split * dataset_size))
np.random.seed(random_seed)
np.random.shuffle(indices)
train_indices, test_indices = indices[split:], indices[:split]

print('Size train set: {}, Size test set: {}'.format(len(train_indices), len(test_indices)))

# Creating PT data samplers and loaders:
train_sampler = SubsetRandomSampler(train_indices)
test_sampler = SubsetRandomSampler(test_indices)

train_loader = DataLoader(dataset, batch_size = batch_size, sampler = train_sampler)
test_loader = DataLoader(dataset, batch_size = batch_size, sampler = test_sampler)
```

```python
class GazeNetwork(nn.Module):
  def __init__(self):
    super(GazeNetwork, self).__init__()
    self.conv1 = nn.Conv2d(4, 32, 4, stride=2, bias = True)
    self.conv2 = nn.Conv2d(32, 64, 3, stride=2, bias = True)
    self.conv3 = nn.Conv2d(64, 64, 5, stride=2, bias = True)
    self.deconv3 = nn.ConvTranspose2d(64,64,5,stride = 2, bias = True)
    self.deconv2 = nn.ConvTranspose2d(64,32,3,stride = 2, bias = True)
    self.deconv1 = nn.ConvTranspose2d(32,1,4,stride = 2, bias = True)
    self.logsoftmax = nn.LogSoftmax(dim = -1)
    
  def forward(self, X, train=True, hard=False):
    X = F.elu(self.conv1(X))
    X = F.elu(self.conv2(X))
    X = F.elu(self.conv3(X))
    X = F.elu(self.deconv3(X))
    X = F.elu(self.deconv2(X))
    X = self.deconv1(X)
    
    return self.logsoftmax(X.view(-1,1,6400)).view(-1,1,80,80)
  
  def load_model(self, path):
    self.load_state_dict(torch.load(path))
```

```python
gazenet = GazeNetwork()
gazenet.load_model('model/saliency_model_5.tar')

criterion = nn.KLDivLoss(size_average = False)
optimizer = optim.Adam(gazenet.parameters(),lr= learn_rate)

```

```python
cor = []
ratio = []
for data in test_loader:
  #data = next(iter(test_loader))
  res = gazenet(data[0])
  res_1 = torch.exp(res.view(-1,6400)).view(-1,1,80,80)
  correct = criterion(res, data[1]).data.numpy()/len(res)
  cor.append(correct)
  r_runs = 20
  random = np.zeros(r_runs)
  for i in range(r_runs):
    shuffle = data[1][torch.LongTensor(np.random.permutation(range(len(data[1]))))]
    r = criterion(res,shuffle)/len(res)
    random[i] = r.data.numpy()
  
  ratio.append(random.mean())
  print('Correct: {:.8f} , Mean random: {:.8f}, Std random: {:.8f}, Ratio: {:.8f}'.format(correct,random.mean(), random.std(), correct/random.mean()))

print('Summary: {:.8f},{:.8f}'.format(np.mean(cor),np.mean(ratio)))
```

```python
# Only execute when you wanna train. :)
for epoch in range(5):  # loop over the dataset multiple times

  running_loss = 0.0
  for i, data in enumerate(train_loader):
    # get the inputs
    inputs, labels = data

    # zero the parameter gradients
    optimizer.zero_grad()

    # forward + backward + optimize
    outputs = gazenet(inputs)
    loss = criterion(outputs, labels)/batch_size
    loss.backward()
    optimizer.step()

    # print statistics
    running_loss += loss.item()
    if i % 100 == 99:    
      print('[{}, {}] loss: {}'.format(epoch + 1, i + 1, running_loss / 100))
      running_loss = 0.0

```
